```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'No.of Ratings Recieved'})
```

```python
ratings
```

**Observations**

1: Rating 4.5 to 4.9=> excellent food

2: Rating 4.0 to 4.4=> very good food

3: Rating 3.5 to 3.9=> good food

4: Rating 2.5 to 3.4=> average food

5: Rating 1.8 to 2.4=> poor food

6: Rating 0: no ratings given




```python
ratings.head()
```

```python
plt.rcParams['figure.figsize']=(20,10)
sns.barplot(x='Aggregate rating',y='No.of Ratings Recieved',data=ratings)
```

```python
# adding colors according to the given data as these are randomly assigned colors
plt.rcParams['figure.figsize']=(20,10)
sns.barplot(x='Aggregate rating',y='No.of Ratings Recieved',data=ratings,hue='Rating color',palette=['grey','red','orange','yellow','green','darkgreen'])
```

**Observations:**

1.   Not rated count is very high
2.   Maximum no.of ratings are between 2.4 to 3.9



```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['grey','red','orange','yellow','green','darkgreen'])
```
